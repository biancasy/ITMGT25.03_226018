```python
import requests
import time

url="http://api.open-notify.org/iss-now.json"
response=requests.get(url)
location=[]

for i in range(100):
    #if response.status_code==200:
    data=response.json()
    latitude=float(data["iss_position"]["latitude"])
    longitude=float(data["iss_position"]["longitude"])
    timestamp=float(data["timestamp"])
    location.append((latitude,longitude,timestamp))
    time.sleep(10)
```

```python
import matplotlib.pyplot as plt

latitudes = [loc[0] for loc in location]
longitudes = [loc[1] for loc in location]

plt.scatter(longitudes, latitudes)
plt.title('Path of the International Space Station')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

#when I ran this, all of the longitudes were 49.5389 and all the latitudes were -115.4175
#thus, even if the graph is supposed to be a line graph that charts through the different longitudes and latitudes, only one big dot is seen
#this is proof that all the longitudes and latitudes for the 100 pings are equal
#it can then be assumed that this is due to the fact that the space station hasn't moved
#once it moves, the graph will show its course based on longitude and latitude


```
